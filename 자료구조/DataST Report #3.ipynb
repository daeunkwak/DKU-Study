```python
class Set :
  def __init__(self) :
    self.items = []
    
  # (1) contains() 연산을 in 연산자를 사용하지 않고 구현
  # item에 해당하는 인덱스를 리턴하도록 구현함
  def contains(self, item) :
      for i in range (len(self.items)) :
          if self.items[i] == item :
              return i
      return False

  # (2) delete() 연산을 remove() 사용하지 않고 다시 구현
  def delete(self, item) :
      if contains(self, item) :
        idx = contains(self, item)
        self.items.pop(idx)

  # (3) insert(), union(), intersect(), difference() 연산도 in 사용하지 않고 구현
  def insert(self, item) :
      Flag = True
      for i in range (len(self.items)) :
          if (self.items[i] == item) :
              Flag = False
      if (Flag == True) :
          self.items.append(item)
          return True
      return False

  def union(self, setB) :
      setC = Set()
      setC.items = list(self.items)
      for elem in setB.items :
          if (self.contains(elem) == False) :
              setC.items.append(elem)
      return setC

  def intersect(self, setB) :
      setC = Set()
      for elem in setB.items :
          if self.contains(elem) :
              setC.items.append(elem)
      return setC

  def difference(self, setB) :
      setC = Set()
      for elem in self.items :
          if (setB.contains(elem) == False) :
              setC.items.append(elem)
      return setC


  # (4) 차집합을 구하는 different() 메소드를 연산자 중복을 이용해 구현
  def __sub__(self, setB) :
      setC = Set()
      for elem in self.items :
          if (setB.contains(elem) == False) :
              setC.items.append(elem)
      return setC
    
  def display(self) :
    print(self.items)
```

# 테스트 

1. 작성된 클래스를 이용하여 두 개의 5차 다항식을 다음과 같이 만드시오. 주어진 리스트는 최대 차수부터 내림차수으로 계수(coefficient)를 표현한 것이며, 프로그램에 맞게 조정하여 입력할 것. 
- f1 = [10, 0, 0, 0, 6, 3]
- f2 = [20, 10, 10, 10, 10, 20]

아래에 코드셀을 만들고, 실행하시오. 

```python
# 추상 자료형으로 정의된 다항식 클래스를 파이썬의 리스트를 이용해 구현하기
class Polynomial :
    def __init__(self):
        self.poly = []

    def read_poly(self):
        self.poly = input().split(' ')
        best = int(input("다항식의 최고 차수를 입력하시오 : "))
        for i in range(best, best - len(self.poly), -1):
            print("x^", i, "의 계수 : ", self.poly[len(self.poly) - i - 1])
        return self.poly

    def display(self):
        for i in range (len(self.poly) - 1) :
          print(float(self.poly[i]), "x^", len(self.poly) - i - 1, " + ", end = "")
        print(float(self.poly[-1]))


    # 다항식의 차수를 반환한다
    def degree(self):
        return len(self.poly) - 1

    # 미지수에 scalar를 넣어 계산한 결과를 반환한다
    def evaluate(self, scalar):
        cal = 0
        for i in range (len(self.poly) - 1, 0, -1) :
            # print("i = ", i)
            cal = cal + (scalar ** i)
            # print("cal = ", cal)
        return cal

    # 두 다항식을 더해 새로운 다항식 객체로 반환
    def add(self, rhs):
        new_poly = Polynomial()
        if len(self.poly) > len(rhs.poly) :
            new_poly.poly = [0] * len(self.poly)
            for i in range (len(self.poly) - 1, len(self.poly) - len(rhs.poly) - 1, -1) :
                new_poly.poly[i] = self.poly[i] + rhs.poly[i]
            for i in range (len(self.poly) - len(rhs.poly)) :
                new_poly.poly[i] = self.poly[i]
        else :
            new_poly.poly = [0] * len(rhs.poly)
            for i in range (len(rhs.poly) - 1, len(rhs.poly) - len(self.poly) - 1, -1) :
                new_poly.poly[i] = self.poly[i] + rhs.poly[i]
            for i in range (len(rhs.poly) - len(self.poly)) :
                new_poly.poly[i] = rhs.poly[i]
        return new_poly

    # 두 다항식을 빼 새로운 다항식 객체로 반환
    def subtract(self, rhs):
        new_poly = Polynomial()
        if len(self.poly) > len(rhs.poly) :
            new_poly.poly = [0] * len(self.poly)
            for i in range (len(self.poly) - 1, len(self.poly) - len(rhs.poly) - 1, -1) :
                new_poly.poly[i] = self.poly[i] - rhs.poly[i]
            for i in range (len(self.poly) - len(rhs.poly)) :
                new_poly.poly[i] = self.poly[i]
        else :
            new_poly.poly = [0] * len(rhs.poly)
            for i in range (len(rhs.poly) - 1, len(rhs.poly) - len(self.poly) - 1, -1) :
                new_poly.poly[i] = self.poly[i] - rhs.poly[i]
            for i in range (len(rhs.poly) - len(self.poly)) :
                new_poly.poly[i] = rhs.poly[i]
        return new_poly

    # 두 다항식을 곱해 새로운 다항식 반환 -> 객체로 반환
    def multiply(self, rhs):
        new_poly = Polynomial()
        if len(self.poly) > len(rhs.poly) :
            new_poly.poly = [0] * len(self.poly)
            for i in range (len(self.poly) - 1, len(self.poly) - len(rhs.poly) - 1, -1) :
                new_poly.poly[i] = self.poly[i] * rhs.poly[i]
            for i in range (len(self.poly) - len(rhs.poly)) :
                new_poly.poly[i] = self.poly[i]
        else :
            new_poly.poly = [0] * len(rhs.poly)
            for i in range (len(rhs.poly) - 1, len(rhs.poly) - len(self.poly) - 1, -1) :
                new_poly.poly[i] = self.poly[i] * rhs.poly[i]
            for i in range (len(rhs.poly) - len(self.poly)) :
                new_poly.poly[i] = rhs.poly[i]
        return new_poly

f1 = Polynomial()
f1.poly = [10, 0, 0, 0, 6, 3]
f1.display()

f2 = Polynomial()
f2.poly = [20, 10, 10, 10, 10, 20]
f2.display()
```
