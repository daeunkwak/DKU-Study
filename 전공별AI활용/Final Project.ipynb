### 1. 지역별 인구수 그래프

```python
# -*- coding: utf-8 -*-

# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib 한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_path = './malgun.ttf' # 폰트파일의 위치
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)

# Excel 데이터를 데이터프레임 변환
df = pd.read_excel('인구수_통계청.xls')

df.columns = ['연도', '2016', '2017', '2018', '2019', '2020']
df = df.drop(index=0, axis=0)
df = df.rename(index={1: '계', 2:'서울', 3:'부산', 4:'대구', 5:'인천', 6:'광주', 
                      7:'대전', 8:'울산', 9:'세종', 10:'경기', 11:'강원', 12:'충북',
                      13:'충남', 14:'전북', 15:'전남', 16:'경북', 17:'경남', 18:'제주',
                      19:'수도권'})

# 지정한 범위의 년도에 해당하는 지정한 지역 값만 선택
col = list(map(str, range(2016, 2021)))
df = df.loc[['서울','부산','대구','인천','광주','대전','울산','세종','경기','강원'
             ,'충북','충남','전북','전남','경북','경남','제주','수도권'], col]

df = df.transpose()

# 스타일 서식 지정
plt.style.use('ggplot')

# 막대 그래프 그리기
df.plot(kind='bar', figsize=(20,10), width=0.7,
        color=['orange', 'green','skyblue','blue', 'yellow', 'red','pink',
               'purple','olive','lime','grey','khaki','steelblue','darkslateblue',
               'deeppink','black','yellowgreen','navy'])

plt.title('지역별 총 인구수', size=30)
plt.ylabel('총 인구 수', size=20)
plt.xlabel('기간', size=20)
plt.ylim(0,30000)
plt.legend(loc='best', fontsize=15)

plt.show()
```

### 2. 지역별 인구 밀도 그래프

```python
# Excel 데이터를 데이터프레임 변환
df = pd.read_excel('인구밀도_통계청.xlsx')

df.columns = ['연도', '2016', '2017', '2018', '2019', '2020']
df = df.drop(index=0, axis=0)
df = df.rename(index={1: '계', 2:'서울', 3:'부산', 4:'대구', 5:'인천', 6:'광주', 7:'대전',
                      8:'울산', 9:'세종', 10:'경기', 11:'강원', 12:'충북', 13:'충남', 14:'전북',
                      15:'전남', 16:'경북', 17:'경남', 18:'제주', 19:'수도권'})

# 지정한 범위의 년도에 해당하는 지정한 지역 값만 선택
col = list(map(str, range(2016, 2021)))
df = df.loc[['서울','부산','대구','인천','광주','대전','울산','세종','경기','강원'
             ,'충북','충남','전북','전남','경북','경남','제주','수도권'], col]

df = df.transpose()

# 스타일 서식 지정
plt.style.use('ggplot')

# 막대 그래프 그리기
df.plot(kind='bar', figsize=(15,7), width=0.7,
        color=['orange', 'green','skyblue','blue', 'yellow', 'red','pink',
               'purple','olive','lime','grey','khaki','steelblue','darkslateblue',
               'deeppink','black','yellowgreen','navy'])

plt.title('지역별 총 인구밀도', size=30)
plt.ylabel('총 인구밀도', size=20)
plt.xlabel('기간', size=20)
plt.ylim(0,20000)
plt.legend(loc='best', fontsize=10)

plt.show()
```
